```python
%%time
import timeit
from itertools import combinations

arr = range(100)

for x in range(1,6):
    stmt = f"list(combinations(arr, {x}))"
    setup = "from itertools import combinations; arr = range(100)"
    time_taken = timeit.timeit(stmt, setup, number=1)  # number parametri kertoo, kuinka monta kertaa koodia suoritetaan
    print(f"Time taken for x={x}: {time_taken} seconds")
```
