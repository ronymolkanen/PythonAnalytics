```python
df5 = pd.crosstab(df['sat_salary'], df['gender'])
df5.index = ['Very unsatisfied', 'Unsatisfied', 'Neutral', 'Satisfied', 'Very Satisfied']
df5.columns = ['Male', 'Female']

# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Using seaborn for better color choices
ax = sns.barplot(data=df5.reset_index(), y='index', x='Male', color='skyblue', label='Male')
ax = sns.barplot(data=df5.reset_index(), y='index', x='Female', color='lightcoral', label='Female')

# Annotating the bars
for p in ax.patches:
    count = int(p.get_width())
    if count > 0:
        x = p.get_x() + p.get_width() + 0.05
        y = p.get_y() + p.get_height() / 2
        ax.annotate(count, (x, y), ha='left', va='center', color='black', fontsize=10, weight='bold')

# Adjusting labels and title
plt.xlabel('Number of respondents', fontsize=12)
plt.ylabel('Satisfaction Level', fontsize=12)
plt.title('Satisfaction to Salary by Gender', fontsize=14)

# Adding legend
plt.legend()

# Adding grid
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Removing spines
sns.despine()

plt.show()
```
